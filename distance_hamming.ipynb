```python
import hashlib as hsh
import random as rd 
import secrets as sc
import string as st
def hamming(a: str, b: str) -> int:
    """Retourne la distance de Hamming entre deux chaînes de même longueur."""
    if len(a) != len(b):
        raise ValueError('Les deux chaînes doivent être de même longueur')
    return sum(1 for x, y in zip(a, b) if x != y)
def random_string(l:int)-> str:
    return ''.join(sc.choice(st.ascii_letters + st.digits + st.ascii_lowercase + st.ascii_uppercase) for _ in range(l))
def change_1(char:str)->str:
    if len(char)==0:
        return random_string(1)
    i=rd.randrange(0,len(char))
    a=sc.choice(st.ascii_letters + st.digits + st.ascii_lowercase + st.ascii_uppercase)
    l=list(char)
    l[i]=a
    return ''.join(i for i in l)
def moy(m):
    return sum(m)/len(m)
def simule(l:int)->int:
    """
    Docstring for simule
    
    :param l: l longueur fixe des chaines
    :type l: int
    :return: la distance entre les deux hashs md5 
    :rtype: int
    """
    L=[]
    for _ in range(100000):
        char1=random_string(l)
        L.append(hamming(hsh.md5(char1.encode()).hexdigest(),hsh.md5(change_1(char1).encode()).hexdigest()))
    return moy(L)
L1=[]
for i in range(10):
    L1.append(simule(i))
print(L1)
```
